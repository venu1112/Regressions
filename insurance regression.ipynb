```python
#linear regression
x=ins['Claims']
y=ins['Payment']
x=np.array(x).reshape(-1,1)
y=np.array(y).reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x=sc_x.fit_transform(x)
sc_y=StandardScaler()
y=sc_y.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,reg.predict(x_train),color='blue')
plt.title("calims vs predicted payments")
plt.xlabel("claims")
plt.ylabel("Pred payments")
plt.show()
```

```python
#mean squared error
from sklearn.metrics import r2_score,mean_squared_error
mean_squared_error(y_test,y_pred)
```

```python
#r square
r2_score(y_test,y_pred)
```

```python
#multi regression
a=ins[['Claims','Insured']]
b=ins['Payment']
a=np.array(a).reshape(-1,2)
b=np.array(b).reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
a_train,a_test,b_train,b_test=train_test_split(a,b,test_size=0.2,random_state=5)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
a_train=sc_x.fit_transform(a_train)
a_test=sc_x.fit_transform(a_test)
sc_y=StandardScaler()
b_train=sc_y.fit_transform(b_train)
b_test=sc_y.fit_transform(b_test)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(a_train,b_train)
```

```python
b_pred=reg.predict(a_test)
```

```python
#mean squared error
from sklearn.metrics import r2_score,mean_squared_error
mean_squared_error(b_test,b_pred)
```

```python
#r square
r2_score(b_test,b_pred)
```
